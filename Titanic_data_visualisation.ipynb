```python
df = sns.load_dataset(('titanic'))
```

# TITANIC 데이터의 기본정보출력


```python
  plt.pie(df['survived'].value_counts(), labels=['no', 'yes'], explode=(0,0.1), startangle=90, shadow=True, autopct='%.1f')
  plt.show()
```

```python
  plt.pie(df['who'].value_counts(), labels=['man', 'woman', 'child'], explode=(0,0.1, 0), startangle=90, shadow=True, autopct='%.1f')
  plt.show()
```

```python
  plt.pie(df['embark_town'].value_counts(), labels=['Southhampton', 'Cherbourg', 'Queenstown'], explode=(0,0, 0), startangle=90, shadow=True, autopct='%.1f')
  plt.show()
```

```python
  plt.pie(df['class'].value_counts(), labels=['Third', 'First', 'Second'], explode=(0,0, 0), startangle=90, shadow=True, autopct='%.1f')
  plt.show()
```

```python
  plt.pie(df['alone'].value_counts(), labels=['True', 'False'], explode=(0,0), startangle=90, shadow=True, autopct='%.1f')
  plt.show()
```

# 성별생존자수와생존율의비교

```python
sns.countplot(x='sex', hue='alive', data=df)
```

```python
sns.countplot(x='class', hue='alive', data=df)
```

```python
pd.crosstab(df['class'], df['alive'], margins=True, margins_name='All')
```

```python
sns.countplot(x='class', hue='sex', data=df)
```

```python
pd.crosstab(df['alive'], columns=[df['class'],df['sex']], margins=True, margins_name='All')
```

```python
sns.catplot(data=df, x='class', y='survived', col='sex', kind='bar')
```
